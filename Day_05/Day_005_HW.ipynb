```python
# Import 需要的套件# Import 
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

觀察有興趣的欄位的資料分佈，並嘗試找出有趣的訊息


```python
print(f'Data size = {app_train.shape}')
print(f'Columns = {app_train.columns}')
app_train.head()
```

```python
app_train.describe()
```

```python
print(f'Mean of AMT_INCOME_TOTAL = {app_train["AMT_INCOME_TOTAL"].mean():.0f}')
print(f'STD of AMT_INCOME_TOTAL = {app_train["AMT_INCOME_TOTAL"].std():.0f}')
```

```python
AMT_INCOME_TOTAL = np.log10(app_train['AMT_INCOME_TOTAL'])
AMT_INCOME_TOTAL.plot(kind='hist', bins=50)
xmin = int(np.floor(AMT_INCOME_TOTAL.min()))
xmax = int(np.ceil(AMT_INCOME_TOTAL.max()))
plt.xlim(xmin, xmax)
plt.xticks(np.arange(xmin, xmax+1), [f'{10**i:.0e}' for i in range(xmin, xmax+1)])
plt.xlabel('AMT_INCOME_TOTAL')
```
