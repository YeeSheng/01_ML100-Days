## 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999


可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train.dtypes.value_counts()
```

```python
app_train.dtypes.isin(['int64', 'float64']).count()
```

```python
# 先篩選數值型的欄位
numeric_columns = app_train.select_dtypes(include=[np.number])
```

```python
print(numeric_columns)
```

```python
# 再把只有 2 值 (通常是 0,1) 的欄位去掉
for col in numeric_columns:
    if len(np.unique(numeric_columns[col])) == 2:
        del numeric_columns[col]

print("Numbers of remained columns:", numeric_columns.shape[1])
```

```python
# 檢視這些欄位的數值範圍
fig, axes = plt.subplots(int(np.ceil(numeric_columns.shape[1] / 2)), 2, figsize=(15, int(np.ceil(numeric_columns.shape[1] / 2)) * 2))
for (col, ax) in zip(numeric_columns.columns.tolist(), axes.ravel()):
    ax.boxplot(numeric_columns[col], vert=False)
    ax.set_yticklabels([])
    ax.set_title(col)
plt.tight_layout()
```

##### 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
x, y = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.plot(x, y)
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
x, y = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.plot(x, y)
plt.xscale('log')
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.show()
```

```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
Your Code Here
"""
x, y = ecdf(app_train['REGION_POPULATION_RELATIVE'])
plt.plot(x, y)

plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
x, y = ecdf(app_train['OBS_60_CNT_SOCIAL_CIRCLE'])
plt.plot(x, y)
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.xlim([-0.5,10]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```
